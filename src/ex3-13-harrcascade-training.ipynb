```python
#test models

import cv2

cap = cv2.VideoCapture(0)
detector = cv2.CascadeClassifier('../models/krungsri-training/cascade.xml')
#out = cv2.VideoWriter('datasets/outputs/detection.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 30, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        img = frame.copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = detector.detectMultiScale(gray, 1.2, 4)

        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
        #out.write(img)
        cv2.imshow('Face Detector', img)

        if cv2.waitKey(1) & 0xFF == ord(' '):
            break
    else:
        break
cap.release()
#out.release()
cv2.destroyAllWindows()
```
